```python
train_data.drop(columns = ["Name" , "Cabin","Ticket"], inplace =  True )
test_data.drop(columns = ["Name" , "Cabin","Ticket"], inplace =  True )
```

```python
test_data.head()
```

```python
train_data.head()
```

```python
train_data['Embarked'] = train_data['Embarked'].fillna('S')
test_data['Embarked'] = test_data['Embarked'].fillna('S')
```

```python
train_data['Age'] = train_data['Age'].fillna(train_data['Age'].mean())
test_data['Age'] = test_data['Age'].fillna(test_data['Age'].mean())
```

```python
train_data['Fare'] = train_data['Fare'].fillna(train_data['Fare'].mean())
test_data['Fare'] = test_data['Fare'].fillna(test_data['Fare'].mean())
```

```python

train_data['Embarked'] = train_data['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)


train_data.head()
```

```python

test_data['Embarked'] = test_data['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)


test_data.head()
```

```python
test_data.isnull().sum()
```

```python
train_data.isnull().sum()
```

```python
X_train = train_data.iloc[:,1:].values
y_train = train_data.iloc[:,0].values
X_test = test_data.iloc[:,:].values
```

```python
labelecnoder=LabelEncoder()
X_train[:,1] = labelecnoder.fit_transform(X_train[:,1] )
```

```python
X_train
```

```python
X_test[:,1] = labelecnoder.transform(X_test[:,1] )
```

```python
X_test
```

```python
Standard =  StandardScaler()
X_train = Standard.fit_transform(X_train)
X_test = Standard.transform(X_test)
```

```python
k_range = list(range(1,31))
weight_options = ["uniform", "distance"]
leaf_size=[10,20,30,40,50,60]
p=[1,2,3,4,5] 
algorithm=['auto', 'ball_tree', 'kd_tree', 'brute']

param_grid = dict(n_neighbors = k_range, weights = weight_options,algorithm= algorithm ,leaf_size = leaf_size , p =p)
#print (param_grid)
knn = KNeighborsClassifier()

random = RandomizedSearchCV(knn, param_grid, cv = 10, scoring = 'accuracy')
random.fit(X_train,y_train)

print (random.best_score_)
print (random.best_params_)
print (random.best_estimator_)
```

```python

n_estimators = [10,100,500,1000]

param_grid = dict(n_estimators = n_estimators )
#print (param_grid)
random = RandomForestClassifier()

grid = GridSearchCV(random, param_grid, cv = 10, scoring = 'accuracy')
grid.fit(X_train,y_train)

print (grid.best_score_)
print (grid.best_params_)
print (grid.best_estimator_)
```

```python
svc = SVC()
svc.fit(X_train, y_train)
acc_svc = round(svc.score(X_train, y_train) * 100, 2)
acc_svc
```

```python
#Best accuracy
tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
acc_svc = round(tree.score(X_train, y_train) * 100, 2)
acc_svc
```

```python
y_test =  tree.predict(X_test)
```

```python
test_data["Survived"] = y_test
```

```python
test_data.drop(columns = ["Sex" , "Age","Pclass","SibSp","Parch","Embarked","Fare"], inplace =  True)
```
